```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_histogram
import matplotlib.pyplot as plt
import numpy as np

simulator = AerSimulator()
print("Ready!")
```

---
## 1. Quantum State Basics

A qubit's state is a 2D complex vector:

$$|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$$

where $|\alpha|^2 + |\beta|^2 = 1$.

- $|\alpha|^2$ = probability of measuring 0
- $|\beta|^2$ = probability of measuring 1

---
## 2. Visualizing Gates on the Bloch Sphere

```python
gates = {
    '|0> (start)': QuantumCircuit(1),
    'X gate (NOT)': None,
    'H gate': None,
    'Y gate': None,
    'Z gate on |+>': None,
    'S gate on |+>': None,
}

qc_x = QuantumCircuit(1); qc_x.x(0)
gates['X gate (NOT)'] = qc_x

qc_h = QuantumCircuit(1); qc_h.h(0)
gates['H gate'] = qc_h

qc_y = QuantumCircuit(1); qc_y.y(0)
gates['Y gate'] = qc_y

qc_z = QuantumCircuit(1); qc_z.h(0); qc_z.z(0)
gates['Z gate on |+>'] = qc_z

qc_s = QuantumCircuit(1); qc_s.h(0); qc_s.s(0)
gates['S gate on |+>'] = qc_s

fig, axes = plt.subplots(2, 3, figsize=(15, 10), subplot_kw={'projection': '3d'})
for ax, (name, qc) in zip(axes.flat, gates.items()):
    sv = Statevector.from_instruction(qc)
    plot_bloch_multivector(sv, ax=ax, title=name)

plt.suptitle("Single-Qubit Gates on the Bloch Sphere", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

---
## 3. Gate Summary

| Gate | Matrix | Effect |
|------|--------|--------|
| **X** (NOT) | $\begin{pmatrix} 0 & 1 \\\\ 1 & 0 \end{pmatrix}$ | Flip |0>  |1> |
| **Y** | $\begin{pmatrix} 0 & -i \\\\ i & 0 \end{pmatrix}$ | Rotate pi around Y |
| **Z** | $\begin{pmatrix} 1 & 0 \\\\ 0 & -1 \end{pmatrix}$ | Flip phase of |1> |
| **H** | $\frac{1}{\sqrt{2}}\begin{pmatrix} 1 & 1 \\\\ 1 & -1 \end{pmatrix}$ | Create superposition |
| **S** | $\begin{pmatrix} 1 & 0 \\\\ 0 & i \end{pmatrix}$ | pi/2 phase |
| **T** | $\begin{pmatrix} 1 & 0 \\\\ 0 & e^{i\pi/4} \end{pmatrix}$ | pi/4 phase |

## 4. Gate Identities

```python
def states_equal(qc1, qc2, label):
    sv1 = Statevector.from_instruction(qc1)
    sv2 = Statevector.from_instruction(qc2)
    equal = sv1.equiv(sv2)
    print(f"{label}: {equal}")
    return equal

# HH = I
qc_hh = QuantumCircuit(1); qc_hh.h(0); qc_hh.h(0)
qc_id = QuantumCircuit(1)
states_equal(qc_hh, qc_id, "HH = I")

# XX = I
qc_xx = QuantumCircuit(1); qc_xx.x(0); qc_xx.x(0)
states_equal(qc_xx, qc_id, "XX = I")

# HZH = X
qc_hzh = QuantumCircuit(1); qc_hzh.h(0); qc_hzh.z(0); qc_hzh.h(0)
qc_just_x = QuantumCircuit(1); qc_just_x.x(0)
states_equal(qc_hzh, qc_just_x, "HZH = X")

# HXH = Z
qc_hxh = QuantumCircuit(1); qc_hxh.h(0); qc_hxh.x(0); qc_hxh.h(0)
qc_just_z = QuantumCircuit(1); qc_just_z.z(0)
states_equal(qc_hxh, qc_just_z, "HXH = Z")
```

---
## 5. Measuring Different States

```python
experiments = {
    '|0>': [],
    '|1> (X)': ['x'],
    '|+> (H)': ['h'],
    '|-> (XH)': ['x', 'h'],
}

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for ax, (name, gate_list) in zip(axes.flat, experiments.items()):
    qc = QuantumCircuit(1, 1)
    for g in gate_list:
        getattr(qc, g)(0)
    qc.measure(0, 0)
    counts = simulator.run(qc, shots=2000).result().get_counts()
    plot_histogram(counts, ax=ax, title=name)

plt.suptitle("Measurement Outcomes for Different States", fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```
