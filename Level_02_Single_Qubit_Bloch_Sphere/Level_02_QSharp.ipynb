## 1. Inspecting Quantum States with DumpMachine

Q#'s `DumpMachine()` lets us peek at the full quantum state (only in simulation!).

```python
%%qsharp

open Microsoft.Quantum.Diagnostics;

operation ExploreGates() : Unit {
    use q = Qubit();
    Message("=== |0> state ===");
    DumpMachine();

    Message("\n=== After X gate (|1>) ===");
    X(q); DumpMachine(); X(q);

    Message("\n=== After H gate (|+>) ===");
    H(q); DumpMachine();

    Message("\n=== After Z on |+> (|->) ===");
    Z(q); DumpMachine();
    Reset(q);
}

ExploreGates()
```

---
## 2. All the Pauli Gates

```python
%%qsharp

operation ShowPauliGates() : Unit {
    use q = Qubit();
    Message("--- X gate on |0> ---");
    X(q); DumpMachine(); Reset(q);

    Message("\n--- Y gate on |0> ---");
    Y(q); DumpMachine(); Reset(q);

    Message("\n--- Z gate on |0> ---");
    Z(q); DumpMachine();
    Message("(Z on |0> = |0>, no visible effect!)");
    Reset(q);

    Message("\n--- Z on |+> = |-> ---");
    H(q); Z(q); DumpMachine(); Reset(q);
}

ShowPauliGates()
```

## 3. Measurement Statistics

```python
%%qsharp

operation MeasureSuperposition(shots : Int) : Unit {
    mutable zeros = 0;
    mutable ones = 0;
    for _ in 0..shots - 1 {
        use q = Qubit();
        H(q);
        let r = M(q);
        if r == Zero { set zeros += 1; }
        else { set ones += 1; }
        Reset(q);
    }
    Message($"Superposition measurement ({shots} shots):");
    Message($"  |0>: {zeros} ({zeros * 100 / shots}%)");
    Message($"  |1>: {ones} ({ones * 100 / shots}%)");
}

MeasureSuperposition(10000)
```

## 4. Gate Identities

```python
%%qsharp

operation ProveIdentities() : Unit {
    let shots = 1000;

    // HH = I
    mutable hhOnes = 0;
    for _ in 0..shots - 1 {
        use q = Qubit();
        H(q); H(q);
        if M(q) == One { set hhOnes += 1; }
        Reset(q);
    }
    Message($"HH = I: Got One {hhOnes} times out of {shots} (should be 0)");

    // XX = I
    mutable xxOnes = 0;
    for _ in 0..shots - 1 {
        use q = Qubit();
        X(q); X(q);
        if M(q) == One { set xxOnes += 1; }
        Reset(q);
    }
    Message($"XX = I: Got One {xxOnes} times out of {shots} (should be 0)");

    // HZH = X
    mutable hzhOnes = 0;
    for _ in 0..shots - 1 {
        use q = Qubit();
        H(q); Z(q); H(q);
        if M(q) == One { set hzhOnes += 1; }
        Reset(q);
    }
    Message($"HZH = X: Got One {hzhOnes} times out of {shots} (should be {shots})");
}

ProveIdentities()
```

## 5. The Bloch Sphere

| Position | State | How to reach from |0> |
|----------|-------|---------|
| North pole | |0> | Do nothing |
| South pole | |1> | Apply X |
| +X axis | |+> | Apply H |
| -X axis | |-> | Apply X then H |

**X gate** = 180 deg around X axis, **Z gate** = 180 deg around Z axis, **H gate** = 180 deg around X+Z axis
