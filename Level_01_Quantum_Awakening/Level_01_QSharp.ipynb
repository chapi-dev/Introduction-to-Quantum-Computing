## 3. Your First Q# Operation: Measure a Qubit

Every qubit starts in state |0⟩. If we measure it immediately, we always get `Zero`.

```python
%%qsharp

operation HelloQuantum() : Result {
    // Allocate a single qubit
    use q = Qubit();

    // Measure it — should always return Zero since qubit starts in |0⟩
    let result = M(q);

    // Reset before deallocation
    Reset(q);

    return result;
}
```

```python
# Run the operation from Python
result = qsharp.eval("HelloQuantum()")
print(f"Measurement result: {result}")
print("As expected: Zero! The qubit starts in |0⟩ and we didn't change it.")
```

```python
%%qsharp

// Run it many times to confirm it's always Zero
operation RunHelloMany() : Unit {
    mutable countZero = 0;
    mutable countOne = 0;

    for _ in 0..999 {
        use q = Qubit();
        let r = M(q);
        if r == Zero {
            set countZero += 1;
        } else {
            set countOne += 1;
        }
        Reset(q);
    }

    Message($"Results over 1000 shots:");
    Message($"  Zero: {countZero}");
    Message($"  One:  {countOne}");
}

RunHelloMany()
```

---
## 4. Make It Quantum — The Hadamard Gate

The **Hadamard gate (`H`)** puts a qubit into **superposition**: an equal mix of |0⟩ and |1⟩.

$$H|0\rangle = \frac{|0\rangle + |1\rangle}{\sqrt{2}} = |+\rangle$$

```python
%%qsharp

operation HelloSuperposition() : Unit {
    mutable countZero = 0;
    mutable countOne = 0;

    for _ in 0..999 {
        use q = Qubit();
        
        // Apply H to create superposition
        H(q);
        
        let r = M(q);
        if r == Zero {
            set countZero += 1;
        } else {
            set countOne += 1;
        }
        Reset(q);
    }

    Message($"Results with Hadamard (1000 shots):");
    Message($"  Zero: {countZero} ({countZero * 100 / 1000}%)");
    Message($"  One:  {countOne} ({countOne * 100 / 1000}%)");
    Message($"\nNotice: roughly 50/50! This is quantum randomness.");
}

HelloSuperposition()
```

### Visualize the Circuit

Q# can show circuit diagrams using the widget library:

```python
from qsharp_widgets import Circuit

# Show the circuit for a single Hadamard + Measurement
circuit = qsharp.circuit("{ use q = Qubit(); H(q); let r = M(q); Reset(q); }")
Circuit(circuit)
```

## 1. What Is Quantum Computing?

### Classical Computing (What We Use Today)

Classical computers process information using **bits** — tiny switches that are either **0** (off) or **1** (on). Everything your computer does — from showing this text to running complex simulations — comes down to manipulating billions of these 0s and 1s.

### Quantum Computing (The New Frontier)

Quantum computers use **qubits** (quantum bits). Unlike classical bits, qubits can exist in a state called **superposition** — they can be 0, 1, or *both at the same time*!

| Property | Classical Bit | Qubit |
|----------|:------------:|:-----:|
| States | 0 or 1 | 0, 1, or both simultaneously |
| Analogy | Light switch (on/off) | Coin spinning in the air |
| When measured | Always definite | Collapses to 0 or 1 |

### Why Does It Matter?

Quantum computers aren't just faster classical computers. They're fundamentally different machines that can solve certain problems exponentially faster:

- **Drug Discovery**: Simulating molecular interactions
- **Cryptography**: Breaking and making codes
- **Optimization**: Finding the best route among millions
- **Machine Learning**: Processing vast datasets
- **Materials Science**: Designing new materials

### Limitations of Classical Computers

Some problems are simply too hard for classical computers:
- Simulating a molecule with 50 atoms requires more memory than exists on Earth
- Factoring a 2048-bit number would take classical computers millions of years
- Searching an unsorted database of N items always takes O(N) time classically

## 3. Your First Quantum Operation: "Hello Quantum World!"

In Q#, quantum programs are written as **operations**. The `%%qsharp` magic command lets us write Q# code directly in Jupyter cells.

### Step 1: A Simple Qubit — Always Returns Zero

Every qubit in Q# starts in the |0> state. If we allocate a qubit, measure it immediately, and return the result, we should always get `Zero`.

```python
%%qsharp

// Our first Q# operation!
// 'use' allocates a fresh qubit (always starts in |0>)
// 'M' measures the qubit
// 'Reset' returns the qubit to |0> before releasing it
operation HelloQuantum() : Result {
    use q = Qubit();    // Allocate one qubit
    let result = M(q);  // Measure it
    Reset(q);           // Clean up
    return result;
}
```

```python
# Run it from Python!
result = qsharp.eval("HelloQuantum()")
print(f"Result: {result}")
print("As expected, we got Zero because we didn't apply any gates!")
```

### Step 2: Run It Multiple Times

Let's run the operation many times to confirm it always gives `Zero`:

```python
%%qsharp

// Run HelloQuantum multiple times and count the results
operation RunHelloMany(shots : Int) : (Int, Int) {
    mutable zeroCount = 0;
    mutable oneCount = 0;

    for _ in 0..shots - 1 {
        let result = HelloQuantum();
        if result == Zero {
            set zeroCount += 1;
        } else {
            set oneCount += 1;
        }
    }

    return (zeroCount, oneCount);
}
```

```python
result = qsharp.eval("RunHelloMany(1000)")
print(f"Results over 1000 runs: {result}")
print("All Zeros, as expected!")
```

### Step 3: Make It Quantum — The Hadamard Gate

Now let's make our qubit truly quantum! The **Hadamard gate (H)** puts a qubit into **superposition** — an equal mix of |0> and |1>.

When we measure a qubit in superposition, we get `Zero` about 50% of the time and `One` about 50% of the time.

```python
%%qsharp

// Apply the Hadamard gate before measuring
operation HelloSuperposition() : Result {
    use q = Qubit();
    H(q);               // Put qubit in superposition!
    let result = M(q);
    Reset(q);
    return result;
}
```

```python
%%qsharp

operation RunSuperpositionMany(shots : Int) : (Int, Int) {
    mutable zeroCount = 0;
    mutable oneCount = 0;

    for _ in 0..shots - 1 {
        let result = HelloSuperposition();
        if result == Zero {
            set zeroCount += 1;
        } else {
            set oneCount += 1;
        }
    }

    return (zeroCount, oneCount);
}
```

```python
result = qsharp.eval("RunSuperpositionMany(1000)")
print(f"Results over 1000 runs: {result}")
print("\nNotice: roughly 50% Zero and 50% One!")
print("This is genuine quantum randomness!")
print("\nCongratulations! You've just run your first Q# quantum program!")
```

### Step 4: Visualize the Circuit

Q# provides tools to visualize the quantum circuit:
