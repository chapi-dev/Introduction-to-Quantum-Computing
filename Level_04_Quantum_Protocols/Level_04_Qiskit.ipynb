```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

simulator = AerSimulator()
print("Ready!")
```

---
## 1. Superdense Coding

**Goal**: Alice sends 2 classical bits to Bob by sending only 1 qubit.

**Protocol**:
1. Create a Bell pair
2. Alice encodes her 2-bit message: `00`=I, `01`=X, `10`=Z, `11`=ZX
3. Alice sends her qubit to Bob
4. Bob applies CNOT + H and measures

```python
def superdense_coding(message):
    qr = QuantumRegister(2, 'q')
    cr = ClassicalRegister(2, 'c')
    qc = QuantumCircuit(qr, cr)

    # Create Bell pair
    qc.h(0); qc.cx(0, 1); qc.barrier()

    # Alice encodes
    if message == '01': qc.x(0)
    elif message == '10': qc.z(0)
    elif message == '11': qc.z(0); qc.x(0)
    qc.barrier()

    # Bob decodes
    qc.cx(0, 1); qc.h(0); qc.measure(qr, cr)
    return qc

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for ax, msg in zip(axes.flat, ['00', '01', '10', '11']):
    qc = superdense_coding(msg)
    counts = simulator.run(qc, shots=1024).result().get_counts()
    plot_histogram(counts, ax=ax, title=f'Sent: {msg}')

plt.suptitle("Superdense Coding", fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

```python
qc = superdense_coding('11')
qc.draw('mpl', style='iqp')
```

## 2. Quantum Teleportation

**Goal**: Transfer an arbitrary quantum state from Alice to Bob.

**Protocol**:
1. Create a Bell pair between Alice and Bob
2. Alice has qubit in state |psi> she wants to send
3. Alice applies CNOT then H, measures both qubits
4. Bob applies corrections based on Alice's 2 classical bits

**Key**: The original state is destroyed (no-cloning theorem).

```python
def teleportation_circuit():
    qr = QuantumRegister(3, 'q')
    cr_alice = ClassicalRegister(2, 'alice')
    cr_bob = ClassicalRegister(1, 'bob')
    qc = QuantumCircuit(qr, cr_alice, cr_bob)

    # Prepare state to teleport: Rx(1.2)|0>
    qc.rx(1.2, 0); qc.barrier()

    # Bell pair between q1 and q2
    qc.h(1); qc.cx(1, 2); qc.barrier()

    # Alice's operations
    qc.cx(0, 1); qc.h(0); qc.barrier()

    # Alice measures
    qc.measure(0, cr_alice[0])
    qc.measure(1, cr_alice[1])
    qc.barrier()

    # Bob's corrections
    with qc.if_test((cr_alice[1], 1)): qc.x(2)
    with qc.if_test((cr_alice[0], 1)): qc.z(2)

    qc.measure(2, cr_bob[0])
    return qc

qc = teleportation_circuit()
qc.draw('mpl', style='iqp')
```

```python
qc = teleportation_circuit()
result = simulator.run(qc, shots=4096).result()
counts = result.get_counts()

bob_counts = {'0': 0, '1': 0}
for outcome, count in counts.items():
    bits = outcome.split(' ')
    bob_bit = bits[0]
    bob_counts[bob_bit] += count

print("Bob's results:")
for k, v in bob_counts.items():
    print(f"  |{k}>: {v} ({v*100/4096:.1f}%)")

# Compare with direct measurement
qc_direct = QuantumCircuit(1, 1)
qc_direct.rx(1.2, 0); qc_direct.measure(0, 0)
direct_counts = simulator.run(qc_direct, shots=4096).result().get_counts()
print("\nDirect Rx(1.2)|0> measurement:")
for k, v in sorted(direct_counts.items()):
    print(f"  |{k}>: {v} ({v*100/4096:.1f}%)")
print("\nTeleportation successful!")
```

## 3. Protocol Comparison

| Protocol | Sends | Pre-shared | Classical Bits | Qubits Sent |
|----------|-------|------------|----------------|-------------|
| **Superdense Coding** | 2 classical bits | 1 Bell pair | 0 | 1 |
| **Teleportation** | 1 quantum state | 1 Bell pair | 2 | 0 |

They are **dual protocols**!
